```python
import requests as re
import os
from IPython.display import JSON
import pandas as pd
import json
```

```python
foursquare_id = os.environ["FOURSQUARE_CLIENT_ID"]
foursquare_secret = os.environ["FOURSQUARE_CLIENT_SECRET"] 
```

```python
location = "Richmond,Canada"
v = "20200731"
radius = "1000"
url = "https://api.foursquare.com/v2/venues/search?near="+location+"&client_id="+foursquare_id+"&client_secret="+foursquare_secret+"&v="+v+"&radius="+radius
```

```python
# dont run again
res = re.get(url)
```

```python
foursquareDATA = res.json()
print(foursquareDATA)
```

```python
foursquareDATA['response']
```

```python
foursquarename = []

for biz in foursquareDATA['response']['venues']:
    foursquarename.append(biz['name'])

print(foursquarename)
```

```python
foursquarevenue = []

for biz in foursquareDATA['response']['venues']:
    if len(biz['categories']) == 0:
        foursquarevenue += ["null"]
    else:
        for category in biz['categories']:
            foursquarevenue.append(category['name'])
           
print(foursquarevenue)
```

```python
foursquarevenueID = []

for biz in foursquareDATA['response']['venues']:
    foursquarevenueID.append(biz['id'])

print(foursquarevenueID)
```

```python
venue_id = "4be6247d910020a1050ed414"
venueURL = "https://api.foursquare.com/v2/venues/"+venue_id+"?"+"&client_id="+foursquare_id+"&client_secret="+foursquare_secret+"&v="+v
res2 = re.get(venueURL)
foursquareVENUEDATA = res2.json()
```

```python
#foursquarerating = []
ratinginfo = []

for venue_id in foursquarevenueID:
    venueURL = "https://api.foursquare.com/v2/venues/"+venue_id+"?"+"&client_id="+foursquare_id+"&client_secret="+foursquare_secret+"&v="+v
    res2 = re.get(venueURL)
    foursquareVENUEDATA = res2.json()
    ratinginfo.append(foursquareVENUEDATA)
    
    #for biz in foursquareVENUEDATA['response']['venue']:
        #foursquarerating.append(biz['rating'])
    #print(foursquareVENUEDATA)
    print(ratinginfo)
```

```python
ratinginfo
```

```python
foursquarerating = []

for info in ratinginfo:
    if 'rating' in info['response']['venue'].keys():
        foursquarerating.append(int(info['response']['venue']['rating']))
    else:
        foursquarerating.append(int("0"))

print(foursquarerating)
```

```python
foursquaredatabase = {}

foursquaredatabase['Name'] = foursquarename
foursquaredatabase['Category'] = foursquarevenue
foursquaredatabase['Venue ID'] = foursquarevenueID
foursquaredatabase['Rating'] = foursquarerating

foursquaredataframe = pd.DataFrame(data = foursquaredatabase)
foursquaredataframe
```

```python
app_id = os.environ["YELP_CLIENT_ID"]
app_secret = os.environ["YELP_CLIENT_SECRET"]
ENDPOINT = 'https://api.yelp.com/v3/businesses/search'
HEADERS = {'Authorization': 'bearer %s' % app_secret}
```

```python
PARAM = {'location':'Richmond, Canada',
        'radius': 1000}

#PARAM = {'location':'Richmond, Canada'}

response = re.get(url = ENDPOINT, params = PARAM, headers = HEADERS)
```

```python
yelpDATA = response.json()
print(yelpDATA)
```

```python
print(yelpDATA.keys())
```

```python
yelpbizname =[]

for biz in yelpDATA['businesses']:
    yelpbizname.append(biz['name'])
print(yelpbizname)
```

```python
yelpbizrating =[]

for biz in yelpDATA['businesses']:
    yelpbizrating.append(int(biz['rating']*2))
print(yelpbizrating)
```

```python
yelpdatabase = {}

yelpdatabase['Name'] = yelpbizname
yelpdatabase['Rating'] = yelpbizrating

yelpdataframe = pd.DataFrame(data = yelpdatabase)
yelpdataframe
```

```python
combodatabase = {}

combodatabase['Name'] = yelpbizname + foursquarename
combodatabase['Rating'] = yelpbizrating + foursquarerating

df = pd.DataFrame(data = combodatabase)
df.sort_values(by= ['Rating'], ascending = False)
df
```
